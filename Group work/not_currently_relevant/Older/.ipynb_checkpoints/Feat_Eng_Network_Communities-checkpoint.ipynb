```python
# In this notebook, we repeat the working from the 'Feat_Eng_Network_Circular', 
# but now create a graph that computes the neighborhoods of features.
# Then, we compute 'total_weights' by summing over each feature's correlations
# with ONLY THOSE FEATURES IN ITS NEIGHBORHOOD, as opposed to the whole 
# feature set.
```

```python
# imports
from community import community_louvain as community
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx 
```

```python
# load data
root_dir = './DataFiles/'
train_file = root_dir + 'CreditCard_train.csv'
test_file = root_dir + 'CreditCard_test.csv'

_df_train = pd.read_csv(train_file, index_col=0, header=1).rename(columns={'PAY_0':'PAY_1'})
_df_test = pd.read_csv(test_file, index_col=0, header=1).rename(columns={'PAY_0':'PAY_1'})
```

```python
# correlation matrix excl. target
df_train_vars = _df_train.iloc[:, :-1]
df_corr = df_train_vars.corr()

# extract feature names for nodes
features = df_corr.index.values

# convert to numpy matrix
corr_matrix = np.asmatrix(df_corr)

# extract correlations with target label
target_corr = _df_train.corr().iloc[-1, :-1].to_numpy()
```

```python
# Creates graph using the data from the correlation matrix
G = nx.from_numpy_matrix(corr_matrix)

# relabels nodes to match features
G = nx.relabel_nodes(G, lambda x: features[x])

# shows the edges with their corresponding weights
G.edges(data=True)
```

```python
# Function to display the basic graph

def create_corr_network_1(G, min_corr):
    
    # create copy
    H = G.copy()
    
    # checks all edges and removes some based on corr_dir, and min_corr
    for feat_1, feat_2, weight in G.edges(data=True):
        if abs(weight['weight']) < min_corr:
            H.remove_edge(feat_1, feat_2)
    
    # create list for the edges and for the weights
    edges, weights = zip(*nx.get_edge_attributes(H,'weight').items())
    
    # increase weight values to make vals more distinct 
    weights = tuple([(1 + abs(x))**2 for x in weights])
    
    # communities: compute best partition
    neighborhoods = community.best_partition(H, random_state=42)
    
    # positions
    positions = nx.spring_layout(H)
    
    # fig size
    plt.figure(figsize=(15,15))
    
    # draws nodes
    nx.draw_networkx_nodes(H, 
                           positions,
                           neighborhoods.keys(),
                           node_color='#DA70D6',
                           node_size=500, 
                           alpha=0.8)
    
    # styling for labels
    nx.draw_networkx_labels(H, 
                            positions, 
                            font_size=8,
                            font_family='sans-serif')
    
    # color setting
    main_color = plt.cm.GnBu
    
    # draws the edges
    nx.draw_networkx_edges(H, 
                           positions, 
                           edgelist=edges, 
                           style='solid',
                           width=weights,
                           edge_color=weights,
                           edge_cmap=main_color,
                           edge_vmin=min(weights),
                           edge_vmax=max(weights))
    
    # displays the graph without axis
    plt.axis('off')
    
    # show
    plt.show()
    
    return neighborhoods
```

```python
# Plot corr network
nbr = create_corr_network_1(G, 0.2)
```

```python
edges = G.edges(data=True)
total_corr = {}

# iterate over features and sum over weights 
# across neighborhood to compute total corr
# for each feature
for feat in features:
    total_corr[feat] = 0
    for feat_i, feat_j, weight in edges:
        if feat_i == feat:
            if feat_i != feat_j and nbr[feat_i] == nbr[feat_j]:
                total_corr[feat] += weight['weight']
                
# multiply total corr for each variable by its correlation with
# the target label
total_weights = {}
for i, feat in enumerate(features):
    total_weights[feat] = total_corr[feat] * abs(target_corr[i])
```

```python
# check
total_weights
```

```python
# pick highest-corr features from each neighborhood and test simple classifier

feats = ['LIMIT_BAL', 'PAY_1', 'BILL_AMT1']

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score

classifier = SVC()

classifier.fit(_df_train[feats], _df_train.iloc[:, -1])
preds = classifier.predict(_df_test[feats])
```

```python
def get_scores(y_test, y_pred):
    scores = dict()
    scores['acc'] = accuracy_score(y_test, y_pred)
    scores['prec'] = precision_score(y_test, y_pred)
    scores['rec'] = recall_score(y_test, y_pred)
    report = "SCORES:\n \
    Accuracy: {:.3f}\n \
    Precision: {:.3f}\n \
    Recall: {:.3f}".format(scores['acc'], scores['prec'], scores['rec'])
    return scores, report
```

```python
# print scores
print(get_scores(_df_test.iloc[:, -1], preds)[1])
```
